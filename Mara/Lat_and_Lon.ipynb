```python
# import directories
import os
import csv
import pandas as pd
import requests
import json

# Google developer API key
from config import gkey

# create path for data file
data_csv = os.path.join('..','cleanData','CleanCountryCodes.csv')

# use read_csv function from pandas directory to create a data frame
data_df=pd.read_csv(data_csv, low_memory = False)
```

```python
def getlatlng (params):
    baseURL = "https://maps.googleapis.com/maps/api/geocode/json"
    response = requests.get(baseURL, params=params)
    data = response.json()
    lat = data['results'][0]['geometry']['location']['lat']
    lng = data['results'][0]['geometry']['location']['lng']
    return lat,lng
```

```python
lat_lng_param = {'address': "Sub-Saharan Africa",
          'key': gkey
         }
getlatlng (lat_lng_param)
```

```python
subRegion = data_df['Sub-region Name'].unique()
subRegion = pd.DataFrame({'Sub-region Name': subRegion,
                        'Lat': "",
                        'Lon': ""
                         })
subRegion = subRegion.dropna(how='any')
```

```python
lat_list = []
lon_list = []
for index in subRegion['Sub-region Name']:
    lat_lng_param = {'address': index,'key': gkey }
    
    lat, lon = getlatlng (lat_lng_param)
    
    lat_list.append(lat)
    lon_list.append(lon)
```

```python
subRegion['Lat'] = lat_list
subRegion['Lon'] = lon_list
subRegion
```

```python
region = data_df['Region Name'].unique()
region = pd.DataFrame({'Region Name': region,
                        'Lat': "",
                        'Lon': ""
                         })
region= region.dropna(how='any')
```

```python
lat_list = []
lon_list = []

for index in region['Region Name']:
    lat_lng_param = {'address': index,'key': gkey }
    
    lat, lon = getlatlng (lat_lng_param)
    
    lat_list.append(lat)
    lon_list.append(lon)
```

```python
region['Lat'] = lat_list
region['Lon'] = lon_list
region
```

```python
country = data_df['Country or Area'].unique()
country = pd.DataFrame({'Country Name': country,
                        'Lat': "",
                        'Lon': ""
                         })
country= country.dropna(how='any')
```

```python
lat_list = []
lon_list = []

for index in country['Country Name']:
    lat_lng_param = {'address': index,'key': gkey }
    
    lat, lon = getlatlng (lat_lng_param)
    
    lat_list.append(lat)
    lon_list.append(lon)
```

```python
country['Lat'] = lat_list
country['Lon'] = lon_list
country
```

```python
# write to csv
subRegion_csv = os.path.join('.','subRegion.csv')
subRegion.to_csv(subRegion_csv, index=False, header=True)

region_csv = os.path.join('.','region.csv')
region.to_csv(region_csv, index=False, header=True)

country_csv = os.path.join('.','country.csv')
country.to_csv(country_csv, index=False, header=True)
```
